```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv("./dataset/BX-Books.csv", sep=";",
                    on_bad_lines="skip", encoding="latin-1")
books.head()
```

```python
books.columns
```

```python
# Removing IMG url clms

# books=books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher']]
books.drop(["Image-URL-S", "Image-URL-M", "Image-URL-L"], axis=1, inplace=True)
books.head()
```

```python
books.rename(columns={"Book-Title": "title", "Book-Author": "author",
             "Year-Of-Publication": "year", "Publisher": "publisher"}, inplace=True)
books.head()
```

```python
users = pd.read_csv("./dataset/BX-Users.csv", sep=";",
                    on_bad_lines="skip", encoding="latin-1")
users.head()
```

```python
users.rename(columns={"User-ID": "user_id",
             "Location": "location", "Age": "age"}, inplace=True)
users.head()
```

```python
ratings = pd.read_csv("./dataset/BX-Book-Ratings.csv", sep=";",
                      on_bad_lines="skip", encoding="latin-1")
ratings.head()
```

```python
ratings.rename(columns={"User-ID": "user_id",
               "Book-Rating": "book_rating"}, inplace=True)
ratings.head()
```

```python
print(books.shape)
print(users.shape)
print(ratings.shape)
```

```python
# Finding out the users have given 200+ retings
print(ratings["user_id"].value_counts().shape)
x = ratings["user_id"].value_counts() > 200
```

```python
# These people has given rating 200+ times
x[x]
```

```python
# Getting index of that folkd
y = x[x].index
y
```

```python
# ratings["user_id"].isin(y) # It returns series
ratings = ratings[ratings["user_id"].isin(y)]
ratings
```

```python
# Merging ratings and books on the basis of ISBN
ratings_with_books = ratings.merge(books, on="ISBN")
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
# Finding the books which having 50+ ratings
# for it we can apply groupby on ISBN or title, here we are applying on title so we can use that fiels later on
number_rating = ratings_with_books.groupby(
    "title")["book_rating"].count().reset_index()
number_rating.head()
```

```python
number_rating.rename(
    columns={"book_rating": "num_of_rating"}, inplace=True)
number_rating.head()
```

```python
# Again merging it with the ratings on the basis of title
final_ratings = ratings_with_books.merge(number_rating, on="title")
final_ratings.head()
```

```python
print(final_ratings.shape)
```

```python
final_ratings = final_ratings[final_ratings["num_of_rating"] >= 50]
final_ratings.head()
```

```python
final_ratings.shape
```

```python
# This ratings are given by the users who have made 200+ ratings and book having 50+ rartings
final_ratings.drop_duplicates(["user_id", "title"], inplace=True)
final_ratings.shape
```

```python
book_pivot = final_ratings.pivot_table(
    columns="user_id", index="title", values="book_rating")
book_pivot
```

```python
book_pivot.fillna(0, inplace=True)
book_pivot
```

```python
# Creating a sparse matrix
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
# algorithm="brute" means calculating distance using each node, by defaults is skips few nodes
model = NearestNeighbors(algorithm="brute")
```

```python
model.fit(book_sparse)
```

```python
# book_pivot.iloc[231, :].values => It gives 237th row and all clms
# after we take that row values so it gives series so we make reshape it and traspose it
# Neighbours = 6, because one would be ir self and we reuired 5
distance, suggestions = model.kneighbors(
    book_pivot.iloc[237, :].values.reshape(1, -1), n_neighbors=6)
```

```python
# It shows 5 closest distance, 0 is itself
distance
```

```python
# It gives it og that book
suggestions
```

```python
# book_pivot.index It gives all the book, and by providing index we'll get that book only
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```
